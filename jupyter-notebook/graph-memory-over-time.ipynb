```python
print("Input file location")
file = input()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv(file)

memory_used = df["MEMORY_USED"]

# Plot the MEMORY_USED column over time
plt.figure(figsize=(10, 6))
plt.plot(memory_used.index, memory_used, marker='o', linestyle='-')
plt.xlabel("Time [s]") 
plt.ylabel("used memory [MB]") 
plt.title(f"used memory over Time for {file}") 
plt.grid() 
plt.xticks(rotation=45) 

# Show the plot
plt.tight_layout()
plt.show()
```
